## ベクトルの類似度比較による文書検索

### 想定質問の回答率

```python
import ast
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from openai import OpenAI
import pandas as pd
```

```python
embedded_df
```

```python
#質問設定
q = q_df["q_embedding"].apply(lambda x: np.array(ast.literal_eval(x)))

#元データの読み込み
embedded_df['test_q_embedding_af'] = embedded_df['test_q_embedding'].apply(lambda x: np.array(ast.literal_eval(x)))

pre_num = []

for i in range(len(q)):
# コサイン類似度を計算
    similarities = cosine_similarity([q[i]], list(embedded_df['test_q_embedding_af']))

    # 類似度が最大の行を取得
    most_similar_index = similarities.argmax()
    most_similar_chunk = embedded_df.loc[most_similar_index, 'test_chunk']
    most_similar_label = embedded_df.loc[most_similar_index, 'label_num']
    # 類似度も取得
    max_similarity = similarities[0, most_similar_index]

    # 結果の出力
    print(f"質問文[{i}]：\n",q_df['user_question'][i])
    print(f"最大類似度:: {max_similarity:.3f}")
    print("最も類似度が高い行のchunk:", most_similar_chunk)
    print("最も類似度が高い行のlabel:", most_similar_label)
    print("＝"*30)
    pre_num.append(most_similar_label)
print(pre_num)
```

```python
len(q_df)
```

```python
Ans = q_df["label_num"]
Ans = Ans.to_numpy()
Ans
```

```python
data_len = len(Ans)

correct_data = 0
for i in range(data_len):
    if pre_num[i] == Ans[i]:
        correct_data += 1

print(correct_data/data_len)
```

## ベクトルの類似度比較による文書検索

### 想定質問の回答率

```python
import ast
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from openai import OpenAI
import pandas as pd
```

```python
#質問設定
pre_q = df_test["q_embedding"].apply(lambda x: np.array(ast.literal_eval(x)))

#元データの読み込み
base_df['q_embedding'] = base_df['q_embedding'].apply(lambda x: np.array(ast.literal_eval(x)))

pre_num = []

for i in range(len(pre_q)):
# コサイン類似度を計算
    similarities = cosine_similarity([pre_q[i]], list(base_df['q_embedding']))

    # 類似度が最大の行を取得
    most_similar_index = similarities.argmax()
    most_similar_chunk = base_df.loc[most_similar_index, 'chunk']
    most_similar_label = base_df.loc[most_similar_index, 'label_num']
    # 類似度も取得
    max_similarity = similarities[0, most_similar_index]

    # 結果の出力
    print(f"質問文[{i}]：\n",df_test['test_question'][i])
    print(f"最大類似度:: {max_similarity:.3f}")
    print("最も類似度が高い行のchunk:", most_similar_chunk)
    print("最も類似度が高い行のlabel:", most_similar_label)
    print("＝"*30)
    pre_num.append(most_similar_label)
print(pre_num)
```

```python
Ans = df_test["label_num"]
Ans = Ans.to_numpy()
Ans
```

```python
data_len = len(Ans)

correct_data = 0
for i in range(data_len):
    if pre_num[i] == Ans[i]:
        correct_data += 1

print(correct_data/data_len)
```

### 類似度の傾向を見る場合

```python
import ast
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# データの読み込み
embedded_df = pd.read_csv('C:/Users/iput Hata Sakuto/cuda-11.8/Univercity/QA/q_embedding_df.csv',encoding='utf-8')  # embedded_df.csv を適切なファイルパスに置き換えてください
embedded_df['q_embedding'] = embedded_df['q_embedding'].apply(ast.literal_eval)

# コサイン類似度を計算
similarities = cosine_similarity([q_emb], list(embedded_df['q_embedding']))[0]

# 類似度の上位5個を取得
top_5_indices = np.argsort(similarities)[-5:][::-1]

# 上位5個の行と類似度を表示
for idx in top_5_indices:
    chunk = embedded_df.loc[idx, 'chunk']
    similarity = similarities[idx]
    print(f"類似度: {similarity:.3f}, \n {chunk}")
```

### 想定質問の作成

```python
from openai import OpenAI
import pandas as pd
import os

api_key = os.environ["OPENAI_API_KEY"]
# OpenAI クライアントの初期化
client = OpenAI(api_key=api_key)
```

```python
import openai
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# dfからchunkデータを取得
base_chunks = df['question'].tolist()

# ChatGPTに言い換えをリクエスト
q_summarize_data = []
total = len(base_chunks)*40
chunk_counter = 0

for i in range(total):
    su_target = base_chunks[chunk_counter]
    print("対象：",chunk_counter)
    if i % 40 == 0:
        if i > 0:
            chunk_counter += 1
        
    while True:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo", 
            messages=[
                {"role": "user", "content": f"次の文章は質問文である。類似の質問文を作成して。: {su_target}"}
            ]
        )
        
        content = response.choices[0].message.content

        # 生成が上手くいかない場合、再生成させる
        if content not in q_summarize_data:
            break

    print(content)
    print("=" * 100)
    q_summarize_data.append(content)
```

```python
# 新しいDataFrameをcsvファイルに書き出す
df_question = pd.DataFrame({
              'test_question': q_summarize_data
              })

df_question.to_csv('C:/Users/iput Hata Sakuto/cuda-11.8/提出/自作モデル-2/0. 想定質問作成/test_question_Q.csv', index=False)

df_question
```

```python
import pandas as pd
file_path_test = "C:/Users/iput Hata Sakuto/cuda-11.8/提出/自作モデル-2/0. 想定質問作成/test_question_Q.csv"
df_test = pd.read_csv(file_path_test, encoding='utf-8')
```

```python
#各questionに対する期待値（ラベル）の設定
ans_label = []
label_count = 0

for label_num in range(len(df_test)):
  ans_label.append(label_count)
  if label_num % 40 == 0 and label_num >0 :
    label_count += 1

print(ans_label)
```

```python
label_list = df["label_num"].tolist()
label_list
```

```python
label_data = []

for target_label in range(len(ans_label)):
  label_data.append(label_list[ans_label[target_label]])
print(label_data)
```

```python
df_test["label_num"] = label_data
```

```python
df_test.loc[df_test["label_num"] == 0, "label"] = "専門職大学"
df_test.loc[df_test["label_num"] == 1, "label"] = "特徴"
df_test.loc[df_test["label_num"] == 2, "label"] = "学び"
df_test.loc[df_test["label_num"] == 3, "label"] = "授業"
```

```python
from openai import OpenAI
import pandas as pd

def pd_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input=[text], model=model).data[0].embedding

# データフレームの各行に対して get_embedding 関数を適用
df_test['q_embedding'] = df_test['test_question'].apply(lambda x: pd_embedding(x, model='text-embedding-ada-002'))

# 最終的にデータフレームを CSV ファイルに保存します
df_test.to_csv("C:/Users/iput Hata Sakuto/cuda-11.8/提出/自作モデル-2/0. 想定質問作成/test_question_Q.csv", index=False)
df_test
```

```python
from openai import OpenAI
import pandas as pd
import os

api_key = os.environ["OPENAI_API_KEY"]
# OpenAI クライアントの初期化
client = OpenAI(api_key=api_key)

# データフレームの取得
file_path = "C:/Users/iput Hata Sakuto/cuda-11.8/提出/ベースライン/out_put/database.xlsx"
df = pd.read_excel(file_path)
df
```

## エクセルデータにembeddingを付与

```python
from openai import OpenAI
import pandas as pd

def pd_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input=[text], model=model).data[0].embedding

# データフレームの各行に対して get_embedding 関数を適用
df['q_embedding'] = df['question'].apply(lambda x: pd_embedding(x, model='text-embedding-ada-002'))

# データフレームの各行に対して get_embedding 関数を適用
df['a_embedding'] = df['chunk'].apply(lambda x: pd_embedding(x, model='text-embedding-ada-002'))

# 最終的にデータフレームを CSV ファイルに保存します
df.to_csv('C:/Users/iput Hata Sakuto/cuda-11.8/提出/自作モデル-2/0. 想定質問作成/q_embedding_df.csv', index=False)
df
```

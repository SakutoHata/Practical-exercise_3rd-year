```python
from openai import OpenAI
import pandas as pd
import os

api_key = os.environ["OPENAI_API_KEY"]
# OpenAI クライアントの初期化
client = OpenAI(api_key=api_key)
```

### 想定質問の作成

```python
detail_order = """
               以下のルールを覚えて遵守してください。\n
               1.文章を略すことなく、全て日本語で返答してください。\n
               2.提供された情報に基づき文章を作成してください。\n
               3.返答時の出力では「以上、50個のQを作成しました」のような本文以外の内容は含めず、文頭に「Q1:」のようなものは付けない。\n
               4.文章の生成数に関して指示がある場合は、途中もすべて表示して。中略や〇〇に続きます、（以下略）などの表記を行わず、全文生成する。\n
               5.返答時に指示語で、AとQを使わないように。
               6.生成文は1つにつき、30文字程度で作成するように。
               7.Qは毎回、必ず70個作成すること。\n「1.」から「70.」まで必ず作成するように。
               """
```

```python
# dfからchunkデータを取得
base_data = df['chunk'].tolist()
refer_q = df["question"].tolist()
# 生成した質問文の格納先
txt_data = []

for i in range(len(base_data)):
  su_target = base_data[i]
  refer_target = refer_q[i]
  response = client.chat.completions.create(
          model="gpt-3.5-turbo",
          messages=[{"role": "system",
                    "content": f"QA文の作成を行っている。次の文章はAである。A:{su_target}"},
                    {"role": "system", "content": f"Qの作成の際、{refer_target}という観点に重きを置いて作成して。"},
                    {"role": "system", "content": detail_order},
                    {"role": "assistant", "content": "ルールを覚えました。次の指示が来た場合、「1.」から「70.」まで、70個生成します。"},
                    {"role": "user",
                    "content": f"Aを参照して短文かつ日本語でAに対応するQを70個作成して。\n生成数：70個 \nQ："}
                  ]
      )
  # responseからcontentを抽出してsu_dataに格納
  content = response.choices[0].message.content

  #生成が上手くいかない場合、再生成させる
  if "70." not in content:
    response = client.chat.completions.create(
          model="gpt-3.5-turbo",
          messages=[{"role": "system",
                    "content": f"QA文の作成を行っている。次の文章はAである。A:{su_target}"},
                    {"role": "system", "content": f"Qの作成の際、{refer_target}という観点に重きを置いて作成して。"},
                    {"role": "system", "content": detail_order},
                    {"role": "assistant", "content": "ルールを覚えました。次の指示が来た場合、「1.」から「70.」まで、70個生成します。"},
                    {"role": "user",
                    "content": f"Aを参照して短文かつ日本語でAに対応するQを70個作成して。\n生成数：70個 \nQ："}
                  ]
      )
    content = response.choices[0].message.content
  else:
    print("Next:",i,)

  print(content)
  print("="*30)
  txt_data.append(content)
```

```python
sentences = []
for l in range(len(txt_data)):
  txt = txt_data[l]
  lines = txt.split('\n')
  sentences += lines

print(sentences)
```

```python
import re

def is_start_with_number(text):
    # 文頭が数字でないかどうかを判定する正規表現
    pattern = re.compile(r"^\d")
    # マッチしたらFalse、それ以外はTrue
    return not bool(pattern.match(text))

check_list = []

for num in range(len(sentences)):
  target_sentence = is_start_with_number(sentences[num])
  #print(target_sentence)
  if target_sentence == False:
    check_list.append(sentences[num])
  else:
    print("None")
print(check_list)
```

```python
import re

def remove_numbers_prefix(sentences):
    # 正規表現を使用して「数字.」を削除
    cleaned_sentences = [re.sub(r'^\d+\.', '', sentence) for sentence in sentences]
    #cleaned_sentences = [re.sub(r'^\d+\:', '', sentence) for sentence in sentences]
    return cleaned_sentences

# 「数字.」を削除した結果を表示
cleaned_txt = remove_numbers_prefix(check_list)
print(cleaned_txt)
```

```python
All_ans = []
All_ans += cleaned_txt
#print(All_ans)
```

```python
#各questionに対する期待値（ラベル）の設定
ans_label = []
label_count = 0

for label_num in range(len(All_ans)):
  ans_label.append(label_count)
  if label_num % 70 == 0 and label_num >0 :
    label_count += 1

#print(ans_label)
```

```python
#解答データの設定
chunk_data = df["chunk"].tolist()
ans_data = []
ans_count = 0

for ans_num in range(len(All_ans)):
  ans_data.append(chunk_data[ans_label[ans_num]])

#print(ans_data)
```

```python
label_list = df["label_num"].tolist()
label_list
```

```python
label_data = []


for target_label in range(len(ans_label)):
  label_data.append(label_list[ans_label[target_label]])
print(label_data)
```

```python
import pandas as pd

q_df = pd.DataFrame({'test_question': All_ans,
                     'test_chunk':ans_data,
                     'label_num':label_data,
                    })
```

```python
q_df.loc[q_df["label_num"] == 0, "label"] = "専門職大学"
q_df.loc[q_df["label_num"] == 1, "label"] = "特徴"
q_df.loc[q_df["label_num"] == 2, "label"] = "学び"
q_df.loc[q_df["label_num"] == 3, "label"] = "授業"
```

```python
from openai import OpenAI
import pandas as pd

def pd_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input=[text], model=model).data[0].embedding

# データフレームの各行に対して get_embedding 関数を適用
q_df['test_q_embedding'] = q_df['test_question'].apply(lambda x: pd_embedding(x, model='text-embedding-ada-002'))
```
